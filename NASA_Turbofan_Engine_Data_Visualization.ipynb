## 반복문을 접목시켜 100개의 데이터셋의 분할 및 저장

#### FD001 데이터셋 분할 및 저장

    위에서도 언급했드시 FD001,2,3,4 모두 train과 test 데이터셋과 RUL 데이터셋으로 구성되어 있다.
    
    특히, 고장 종류와 운행 조건에 따라 분류한 FD00N 데이터셋은 아래와 같이 손으로 직접 분할하기에는 너무 많은
    대수의 엔진으로 구성되어 있다.

    FD001 엔진 대수 = train 100 대 + test 100 대 = 200
    FD002 엔진 대수 = train 260 대 + test 260 대 = 520
    FD003 엔진 대수 = train 100 대 + test 100 대 = 200
    FD004 엔진 대수 = train 249 대 + test 249 대 = 498

    총 200 + 520 + 200 + 498 = 1298 대에서 발생한 빅데이터이다. 

    이러한 방대한 데이터를 엑셀을 이용하되 전부 사람이 직접 손으로 데이터를 분할하고 저장하기에는 너무 비효율적이다. 
    
    따라서 분석의 효율을 위해 반복문을 사용해서 FD001,2,3,4의 train과 test를 구성하는 각각의 엔진에 대한 데이터 분할 작업은
    파이썬을 통한 분할 작업 자동화는 필수이다.    
    
    아래의 코드는 FD001의 train 데이터의 합쳐져 있는 데이터를 각각의 엔진으로 분할하는 코드이다.

```python

for i in range(100):

    # 분할된 데이터 파일 저장경로

    folder = 'C:/Users/jhj/Desktop/NASA_Turbofan_Engine_excel_data/엔진 분리형 데이터 모음/train/FD001/'

    Number = str(i+1) + '번 엔진'  # Number라는 변수에 엔진번호를 저장

    Dev_Eng_Data = folder + Number  # 파일이름 지정


    # 1) 쓰기 엔진과 엑셀 파일을 지정해 ExcelWriter 객체생성

    N_th_Eng_Data = pd.ExcelWriter(Dev_Eng_Data, engine = 'xlsxwriter')


    # 2) 데이터를 생성한 엑셀 객체에 쓰기 (시트이름과 index 옵션 지정)

    Devided_Engine_Data = train_FD001.loc[[i+1]]

    Devided_Engine_Data.to_excel(N_th_Eng_Data, sheet_name = 'Sheet1', index = True)


    #3) 객체를 닫고 엑셀 파일로 저장

    N_th_Eng_Data.save()

    print('생성한 파일 : ', Dev_Eng_Data)
```

```python
# 데이터 불러오기

folder = 'C:/Users/jhj/Desktop/NASA_Turbofan_Engine_excel_data/엔진 분리형 데이터 모음/train/FD004/'

Engine_4 = folder + '4번 엔진'


# 1) 쓰기 엔진과 엑셀 파일을 지정해 ExcelWriter 객체생성

Eng4_data = pd.ExcelWriter(Engine_4, engine = 'xlsxwriter')


# 2) 데이터를 생성한 엑셀 객체에 쓰기 (시트이름과 index 옵션 지정)

Devided_Engine_Data = train_FD004.loc[[4]]

Devided_Engine_Data.to_excel(Eng4_data, sheet_name = 'Sheet1', index = True)

#3) 객체를 닫고 엑셀 파일로 저장

Eng4_data.save()

Devided_Engine_Data


# (1) 판다스 Excel Writer 객체 생성

folder = 'C:/Users/jhj/Desktop/NASA_Turbofan_Engine_excel_data/엔진 분리형 데이터 모음/train/FD004/'

Engine_4 = folder + '4번 엔진'

excel_writer = pd.ExcelWriter(Engine_4, engine = 'xlsxwriter')


# (2) DataFrame 데이터를 지정된 워크시트(sheet_name=시트_이름)에 쓰기

worksheet_name = 'Sheet1' # 워크시트 이름을 변수에 할당

Devided_Engine_Data.to_excel(excel_writer, sheet_name = worksheet_name, index = False)

# (3) ExcelWriter 객체에서 워크북(workbook)과 워크시트(worksheet) 객체생성

workbook = excel_writer.book

worksheet = excel_writer.sheets[worksheet_name]

# (4) 차트 객체 생성(원하는 차트의 종류 지정)

chart = workbook.add_chart({'type':'line'}) # 꺾은선형 차트

# (5) 차트를 생성하기 위한 데이터 값의 범위 지정

columns_len = len(Devided_Engine_Data.columns) # df.shape[1]도 가능

for k in range(columns_len -1):
    
    start_row = 1                      # 범위의 시작 행 번호

    start_column = k + 1                # 범위의 시작 열 번호

    end_row = len(Devided_Engine_Data.index)            # 범위의 끝 행 번호 

    end_column = k + 1                 # 범위의 끝 열 번호

    # 그래프 및 축 값을 위한 데이터 범위, 범례 이름, 마커 지정

    chart.add_series({

        'values':[worksheet_name, start_row, start_column, end_row, end_column], # 꺾은선형 차트를 위한 데이터 범위 지정

        'categories':[worksheet_name, start_row, 0, end_row, 0],                 # x 축 값에 사용할 데이터 범위 지정

        'name': [worksheet_name, 0, k + 1],                                      # 범례 이름에 사용할 데이터 지정

        'marker': {'type':'circle'}                                              # 마커 추가 : 원모양(circle)으로 지정

    })


# (5-1) 엑셀 차트에 차트 제목과 x,y 축 제목 추가

chart.set_title({'nmae': 'Train_FD004 No.9 Turbofan Engine Data'})   # 차트 제목 지정

chart.set_x_axis({'name':'Cycle'})                                   # x축 제목 지정

chart.set_y_axis({'name': ''})                                       # y축 제목은 미지정


# (6) 워크시트에 차트가 들어갈 위치 지정해 차트 넣기

worksheet.insert_chart(1, columns_len, chart, {'x_offset' : 25, 'y_offset' : 10})


# (7) ExcelWriter 객체를 닫고 엑셀 파일을 출력

excel_writer.save()

print('생성 파일:', Engine_4)                                # 생성한 파일 이름 출력
```
